Essayer le KNNImputer pour remplecer les NaN


```python
from sklearn.impute import KNNImputer
from sklearn.model_selection import train_test_split
```

```python
X = df.copy().drop(columns=['Country', 'Status', 'Life expectancy'])
y = df['Life expectancy']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
```

```python
# Création de l'imputer avec k=3 voisins
imputer = KNNImputer(n_neighbors=3)

# Appliquer l'imputation
X_train_imp = pd.DataFrame(imputer.fit_transform(X_train), columns=X_train.columns)

X_test_imp = pd.DataFrame(imputer.transform(X_test), columns=X_test.columns)

print(X_train_imp)
```

Corrigeons les oultliers avec np.log1p()

```python
cols_to_transform = ['infant deaths', 'percentage expenditure', 'Measles', 'under-five deaths', 'GDP', 'Population', 'thinness  1-19 years', 'thinness 5-9 years']

X_train_imp[cols_to_transform] = X_train_imp[cols_to_transform].apply(np.log1p)
X_test_imp[cols_to_transform] = X_test_imp[cols_to_transform].apply(np.log1p)
```

```python
from scipy import stats
```

```python
cols_to_box = ['Hepatitis B', 'Polio', 'Diphtheria', 'HIV/AIDS']

X_train_imp[cols_to_box] = X_train_imp[cols_to_box].apply(lambda x: stats.boxcox(x + 1)[0])
X_test_imp[cols_to_box] = X_test_imp[cols_to_box].apply(lambda x: stats.boxcox(x + 1)[0])
```

Faisons les encodages pour les features categorielles

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
enc_train, enc_test = train_test_split(df['Status'], test_size=0.2, random_state=13)
```

```python
enc_train = enc_train.values.reshape(-1, 1)
enc_test = enc_test.values.reshape(-1, 1)
```

```python
encoder = OneHotEncoder()
enc_t = encoder.fit_transform(enc_train)
enc_v = encoder.transform(enc_test)
```

```python
enc_t
```

```python
freq_train, freq_test = train_test_split(df['Country'], test_size=0.2, random_state=13)
```

```python
# Calculer la fréquence de chaque pays
country_count_train = freq_train.value_counts()
country_count_test = freq_test.value_counts()
```

```python
# Mapper la fréquence à chaque pays dans la colonne
freq_train = freq_train.map(country_count_train).values.reshape(-1, 1)
freq_test = freq_test.map(country_count_test).values.reshape(-1, 1)
```

```python
freq_train
```

```python
X_train_new = np.column_stack((scaled_train, enc_t, freq_train)) #enlever enc _ t
X_val_new = np.column_stack((scaled_test, enc_v, freq_test)) #enlever enc_v
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from mlxtend.evaluate import bias_variance_decomp
from xgboost import XGBRegressor
from sklearn.base import clone
from scipy.stats import randint, uniform
from sklearn.model_selection import cross_val_score
```

```python
xgb = XGBRegressor()

param_dist = {
    'n_estimators': randint(50, 500),
    'learning_rate': uniform(0.01, 0.1),
    'max_depth': randint(4, 6),
    'subsample': uniform(0.6, 0.4),
    'colsample_bytree': uniform(0.6, 0.4),
    'gamma': uniform(0, 2),
    'reg_alpha': uniform(0.1, 1),
    'reg_lambda': uniform(0.1, 2)
}

# Initialiser RandomizedSearchCV
random_search = RandomizedSearchCV(estimator=xgb, param_distributions=param_dist,
                                   n_iter=50, cv=5, n_jobs=-1, verbose=2, scoring='neg_mean_squared_error', random_state=13)

# Entraîner le modèle
random_search.fit(X_train_new, y_train)

# Meilleurs paramètres
best_params = random_search.best_params_
print("Meilleurs paramètres trouvés :", best_params)
```

Creation du veritable model

```python
model = XGBRegressor(
    colsample_bytree=best_params["colsample_bytree"],
    gamma=best_params["gamma"],
    learning_rate=best_params["learning_rate"],
    max_depth=best_params["max_depth"],
    n_estimators=best_params["n_estimators"],
    reg_alpha=best_params["reg_alpha"],
    reg_lambda=best_params["reg_lambda"],
    subsample=best_params["subsample"]
)

model.fit(X_train_new, y_train)
```

```python
# Calcola l'R² sul training set
from sklearn.metrics import r2_score

y_train_pred = random_search.predict(X_train_new)
r2_train = r2_score(y_train, y_train_pred)
print("R2 su train: ", r2_train)

# Calcola l'R² sul validation set
y_val_pred = random_search.predict(X_val_new)
r2_val = r2_score(y_test, y_val_pred)
print("R2 su validazione: ", r2_val)
```

```python
mse_train = mean_squared_error(y_train, y_train_pred)
mse_val = mean_squared_error(y_test, y_val_pred)

print("MSE su train: ", mse_train)
print("MSE su validazione: ", mse_val)
```

```python
# Conversion des cibles en arrays NumPy
y_train_np = y_train.to_numpy()
y_val_np = y_test.to_numpy()

# Calcul du biais et de la variance
avg_expected_loss, avg_bias, avg_var = bias_variance_decomp(
    model,
    X_train_new,
    y_train_np,
    X_val_new,
    y_val_np,
    loss='mse',
    random_seed=13
)

print("Average expected loss (MSE): %.3f" % avg_expected_loss)
print("Average bias: %.3f" % avg_bias)
print("Average variance: %.3f" % avg_var)
```
